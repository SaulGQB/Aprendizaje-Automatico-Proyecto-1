### Eliminación de outliers

Como se mostró anteriormente, en la base de datos existen datos anormales u outliers, los cuales pueden tener efectos negativos en la implementación de los modelos de regresión lineal y la ecuación normal, por lo que estos deben removerse del conjunto de datos. Para esto se programaron dos funciones, una que se encarga de identificarlos, a través del primer y tercer cuantil (método IQR), y otra que se encargará de removerlos del dataframe aún sin tratar (este proceso se lleva a cabo antes de la normalización, pues para ella se utilizan los valores mínimos y máximos). 
La función para identificarlos es ``outliers_detect()``, la funión para removerlos es ``remove_outliers()``, ambas funciones de encuentran en el archivo **outliers_treatment.py**.

* ``outliers_detect()`` detecta los outliers a partir del primer y tercer cuantil, que se utilizan para encontrar los límites inferior y superior (bigotes), esta regresa los índices del dataframe donde se encuentran estos valores anormales.

* ``remove_outliers()`` remueve los filas donde se encuentran los outliers utilizando el método drop() de pandas. 

```python
from outliers_treatment import outliers_detect
index_list = []
index_list.extend(outliers_detect(df, 'charges'))
len(index_list)
```

```python
from outliers_treatment import remove_outliers
df = remove_outliers(df, index_list)
len(df)
```

### Transformando las columnas con datos tipo _object_ a valores de 1 y 0

Para utilzar los datos de tipo objeto en los modelos de predicción se realizará la transformación de estos a valores binarios a través del método ``pandas.get_dummies()``, esto se aplicó a las columnas "_sex_", "_smoker_" y "_region_", obteniendo como resultado de aplicar este método los dataframes **genero**, **fumadores** y **region**, con 2, 2 y 4 columnas respectivamente, correspondientes a los diferestes _strings_ de cada columna, por ejemplo, el dataframe **genero** posee las columnas "_male_" y "_female_".

Posteriormente se concatenan los dataframes generados con el dataframe original utilizando la funión ``df.concat()``, (notese que para **genero** y **fumadores** unicamente se concatena una columna, pues solo con esta se obtienen los valores binarios necesarios) y removiendo las columnas con datos de tipo objeto con la función ``df.drop()``.

Por último se cambian los nombres de las columnas para un tratamiento más fácil e intuitivo utilizando la función ``df.rename()``.

```python
genero = pd.get_dummies(df['sex'])
fumadores = pd.get_dummies(df['smoker'])
region = pd.get_dummies(df['region'])
df = pd.concat([df, genero['male'], fumadores['yes'], region], axis=1)
df = df.drop(['sex', 'smoker', 'region'], axis=1)
df = df.rename(columns={'male':'sex', 'yes':'smoker'})
```

### Normalizando datos

Para la normalización de los datos, y con el fin de no utilizar momentaneamente la librería sklearn, se creo la función ``min_max()``, la cual normaliza los datos de 0 a 1 utilizando los valores mínimos y máximos de cada columna de datos. Así mismo, se creo la función ``denormalize()`` para desnormalizar los datos en caso de ser necesario, sin embargo, ya que únicamente los features fueron normalizados no fue necesario su uso (se usaría normalmente para desnormalizar los datos predichos y presentarlos). 

```python
def min_max(df):
    import pandas as pd
    df_min = df.min()
    df_max = df.max()
    df_norm = (df - df_min) / (df_max - df_min)
    df_norm = pd.DataFrame(df_norm)
    return df_norm, df_min, df_max

def denormalize(x_norm, min_x, max_x):
    x = x_norm * (max_x - min_x) + min_x
    return x
```

Realizando pruebas se determinó que es mejor normalizar únicamente los features y no los targets.

```python
X_norm, x_min, x_max = min_max(X)
# Y_norm, Y_min, Y_max = min_max(Y)
```

## Implementación de los modelos

### Ecuación normal

Primero se carga la función ``normal_equation()`` desde el archivo **multivariate_linear_regression.py** (puede encontrarse en este mismo directorio).

```python
from multivariate_linear_regression import normal_equation
```

Se establece el porcentaje de datos para el test como 20% y se aplica la función, introduciendo los datos normalizados "_X__norm_" y "_Y_".

```python
percent = 20 
theta, test_index, train_index, Y_predict = normal_equation(X_norm, Y, percent)
```

```python
print(f'Numero de datos utilizados para el test: {len(test_index)}')
print(f'Numero de datos utilizados para el entrenamiento: {len(train_index)}')
```

Una vez aplicada la función se realizó la gráfica de disperción entre los precios reales y los precios predichos con el modelo de ecuación normal. Puede observarse cierta tendencia en la mayoría de los datos, sin embargo existen algunos que se salen por completo de esta. 

```python
y_test = np.c_[Y.iloc[test_index]]
plt.scatter(Y_predict, y_test)
plt.title('Gráfica de dispersión', fontsize = 14)
plt.xlabel('Y predict', fontsize = 12)
plt.ylabel('Y test', fontsize = 12)
plt.show()
```

Para evaluar el modelo se presentan los indicadores MAE, MAPE y R² , para los primeros dos se espera sean lo más pequeño posible (entre 0 y 1 para el MAPE), mientras que para R² se espera que esta sea lo más cercano a 1 posible. Para este caso los valores optimos junto con sus respectivas $\theta$ se presentan a continuación.

_Los resultados mostrados pueden variar según se ejecute el modelo_

```python
from sklearn import metrics

print(f'MAE: {metrics.mean_absolute_error(y_test, Y_predict)}')
print(f'MAPE: {metrics.mean_absolute_percentage_error(y_test, Y_predict)}')
print(f'R²: {round(metrics.r2_score(y_test, Y_predict), 2)}')
```

### Batch Gradient Descent

Primero se carga la función ``batch_gradient_descent()`` desde el archivo **multivariate_linear_regression.py** (puede encontrarse en este mismo directorio).

```python
from multivariate_linear_regression import batch_gradient_descent
```

Se establece el porcentaje de datos para el test del 20%, las épocas (iteraciones) de 1000, y el coeficiente de aprendizaje $\eta\ de 0.2, de igual manera se introducen los dataframes de features normalizados y el target.

```python
percent = 20
epochs = 1000
eta = 0.2
theta, test_index, train_index, Y_predict, J_log = batch_gradient_descent(X_norm, Y, eta, epochs, percent)
Y_predict[1:5]
```

En la siguiente gráfica se presenta la disminución de la función costo respecto al número de época (iteración) que se realiza internamente en la función  ``batch_gradient_descent()``. El mínimo se encuentra al rededor de la época 100 con $\eta = 0.2$. 

```python
plt.plot(np.arange(epochs), J_log, 'b-', linewidth=2)
plt.xlabel('Epochs')
plt.ylabel(r'$J(\theta)$')
plt.show()
```

Al igual que en el caso anterior, la gráfica de dispersión presenta cierta tendencia lineal con algunos datos dispersos.

```python
y_test = np.c_[Y.iloc[test_index]]
plt.scatter(Y_predict, y_test)
plt.title('Gráfica de dispersión', fontsize = 14)
plt.xlabel('Y predict', fontsize = 12)
plt.ylabel('Y test', fontsize = 12)
plt.show()
```

### Stochastic Gradiente Descent

Primero se carga la función  ``stochastic_gradient_descent()`` desde el archivo **multivariate_linear_regression.py** (puede encontrarse en este mismo directorio).

```python
from multivariate_linear_regression import stochastic_gradient_descent
```

A diferencia de la función ``batch_gradient_descen()`` para este caso se agrega un argumento más, el cual es batch_size, que define el número de datos que se evaluan con cada iteración, al ser stochastic gradient descent el tamaño es de 1.

```python
percent = 20
epochs = 20000
eta = 1
batch_size = 1
theta, test_index, train_index, Y_predict, J_log = stochastic_gradient_descent(X_norm, Y, eta, epochs, percent, batch_size)
Y_predict[1:5]
```

En la siguiente gráfica se presenta la disminución de la función costo respecto al número de época (iteración). En este caso al no trabajar con la totalidad de los datos la función no decrece de manera constante, sino que va cambiando de acuerdo al conjunto de datos utilizados en cada iteración.

```python
plt.plot(np.arange(epochs), J_log, 'b-', linewidth=2)
plt.xlabel('Epochs')
plt.ylabel(r'$J(\theta)$')
plt.show()
```

Al igual que en los casos anteriores, la gráfica de dispersión presenta cierta tendencia lineal con algunos datos dispersos.

```python
y_test = np.c_[Y.iloc[test_index]]
plt.scatter(Y_predict, y_test)
plt.title('Gráfica de dispersión', fontsize = 14)
plt.xlabel('Y predict', fontsize = 12)
plt.ylabel('Y test', fontsize = 12)
plt.show()
```

Para evaluar el modelo se utilizaron los indicadores MAE, MAPE y R². Los valores optimos junto con sus respectivas 𝜃 se presentan a continuación.

_Los resultados mostrados pueden variar según se ejecute el modelo_.
